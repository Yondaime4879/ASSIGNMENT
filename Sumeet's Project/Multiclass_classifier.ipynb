```python
import pandas as pd
data = pd.read_csv("category_headlines.csv")
from sklearn.model_selection import train_test_split
train, test = train_test_split(data, test_size=0.2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.naive_bayes import MultinomialNB
```

```python
################NAIVE_BAYES#####################

from sklearn.pipeline import Pipeline
NB_clf = Pipeline([('vect', CountVectorizer()),('tfidf', TfidfTransformer()),('clf', MultinomialNB())])
NB_clf = NB_clf.fit(train['Sentence'], train['Category'])
```

```python
predicted = NB_clf.predict(test['Sentence'])
```

```python
import numpy as np
np.mean(predicted == test['Category'])
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
########################RANDOM_FOREST##################################
from sklearn.pipeline import Pipeline
RF_clf = Pipeline([('vect', CountVectorizer()),('tfidf', TfidfTransformer()),('clf', RandomForestClassifier(n_estimators=100,max_depth=50))])
RF_clf = RF_clf.fit(train['Sentence'], train['Category'])
```

```python
predicted = RF_clf.predict(test['Sentence'])
er = np.mean(predicted == test['Category'])
print er
# from sklearn import metrics
# fpr, tpr, thresholds = metrics.roc_curve(test['Category'], predicted, pos_label=4)
# A = metrics.auc(fpr, tpr)
# print A
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
########################NEURAL_NETS_Perceptron##################################
from sklearn.pipeline import Pipeline
NN_clf = Pipeline([('vect', CountVectorizer()),('tfidf', TfidfTransformer()),('clf', MLPClassifier(hidden_layer_sizes=(30, ),max_iter=50))])
NN_clf = NN_clf.fit(train['Sentence'], train['Category'])
```

```python
predicted = NN_clf.predict(test['Sentence'])
er = np.mean(predicted == test['Category'])
print er
```
